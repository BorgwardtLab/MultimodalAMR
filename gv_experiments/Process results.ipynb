```python
import pandas as pd
import numpy as np
import os

from os.path import exists, join
import json
```

```python
folder = "outputs/Species_baseline/random_split/SpB_randsplit_driamsC/lowLR_long_BN/metrics"
dataset = "C"
experiment = "SpB_randsplit"
```

```python
metrics = []
for seed in range(100):
    if not exists(join(folder, f"test_metrics_{seed}.json")):
        continue
    with open(join(folder, f"test_metrics_{seed}.json"), "r") as f:
        m = json.load(f)
        
    del m["test_loss"]
    metrics.append(m)
    if len(metrics)==10:
        break
metrics = pd.DataFrame(metrics)
metrics
```

```python
summary_df = pd.concat([metrics.mean(), metrics.std()], axis=1)
summary_df.columns = ["mean", "std"]
summary_df
```

```python
summary_df.to_csv(f"outputs/metrics_results/summary_DRIAMS-{dataset}_{experiment}.csv")
```

# Zero shot

```python
folder = "outputs/MLP_zero_shot_driamsD/lowLR_long_BN_zs/metrics"
dataset = "D"
experiment = "zero_shot"
```

```python
long_table = pd.read_csv("../processed_data/DRIAMS_combined_long_table.csv")
long_table = long_table[long_table["dataset"]==dataset]

drugs_list = sorted(long_table["drug"].unique())
len(drugs_list)
```

```python
zs_metrics = []
for i, drug_name in enumerate(drugs_list):
    try:
        with open(os.path.join(folder, "test_metrics_{}.json".format(i)), "r") as f:
            m = json.load(f)
            m["drug"] = drug_name
    except:
        continue
    zs_metrics.append(m)
zs_df = pd.DataFrame(zs_metrics)

drugs_to_remove = []
for dr in drugs_list:
    df = long_table[long_table["drug"]==dr]
    if len(df["response"].value_counts())<2:
        drugs_to_remove.append(dr)
        
zs_df = zs_df[~zs_df["drug"].isin(drugs_to_remove)].dropna()
zs_df.set_index("drug")
zs_df.to_csv(join("outputs/metrics_results/zero_shot","DRIAMS-{}.csv".format(dataset)), index=False)
zs_df
```
