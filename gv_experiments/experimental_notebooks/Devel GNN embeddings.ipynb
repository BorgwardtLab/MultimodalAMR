```python
import pandas as pd
import json
import numpy as np
```

```python
drugs_df = pd.read_csv("../../processed_data/drug_fingerprints.csv")
drugs_df
```

```python
len(drugs_df.iloc[0]["MACCS_fp"])
```

```python
with open("../../processed_data/GNN_AMRPred_embeddings.json", "r") as f:
    gnn_embeddings = json.load(f)
    
```

```python
len(list(gnn_embeddings.keys()))
```

```python
embeddings = []
for dr in drugs_df["drug"]:
    emb = gnn_embeddings[dr]
    embeddings.append(emb[0])

embeddings = pd.DataFrame(embeddings, index=list(drugs_df["drug"]), columns=[f"z{i}" for i in range(len(emb[0]))])
embeddings
```

```python
embeddings.index.name="drug"
embeddings = embeddings.reset_index()
```

```python
embeddings.to_csv("../../processed_data/GNN_embeddings.csv", index=False)
```
