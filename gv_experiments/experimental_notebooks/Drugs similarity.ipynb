```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

```python
drugs_df = pd.read_csv("../processed_data/drug_fingerprints.csv")
drugs_df
```

```python
drugs_list = sorted(drugs_df["drug"].unique())
drugs_list
```

```python
fp1 = drugs_df["MACCS_fp"].iloc[0]
fp2 = drugs_df["MACCS_fp"].iloc[1]

print(fp1)
print(fp2)
```

```python
fp1 = int(fp1, base=2) # we should be ints
fp2 = int(fp2, base=2) # we should be ints
c = fp1 | fp2
d = fp1 & fp2 
```

```python
for fp_class in tqdm(drugs_df.columns[1:]):
    
    sim_matrix = np.zeros((len(drugs_list), len(drugs_list)))
    
    for i in tqdm(range(len(drugs_list))):
        drug1 = drugs_list[i]
        fp1 = drugs_df[fp_class].iloc[i]
        fp1 = int(fp1, base=2)
        for j in range(i, len(drugs_list)):
            if i==j:
                sim_matrix[i, j] = 1.0
                continue
            drug2 = drugs_list[j]
            fp2 = drugs_df[fp_class].iloc[j]
            fp2 = int(fp2, base=2)
            
            intersection = fp1 & fp2 
            union = fp1 | fp2
            
            tanimoto_index = sum([c=="1" for c in f"{intersection:b}"])/sum([c=="1" for c in f"{union:b}"])
            sim_matrix[i, j] = tanimoto_index
            sim_matrix[j, i] = tanimoto_index
            
    sim_matrix = pd.DataFrame(sim_matrix, index=drugs_list, columns=drugs_list)
    sim_matrix.to_csv("outputs/drugs_similarities/{}_tanimoto_index.csv".format(fp_class))
    
```
