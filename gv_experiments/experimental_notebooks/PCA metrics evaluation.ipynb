```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

import json
```

```python
splits = ["random", "partitioned", "drugs_zero_shot"]
datasets = ["A", "B", "C", "D"]
```

# Joint PCA

```python
template = "JointPCA_DRIAMS-{}_{}_metrics"


joint_pca_results = []
for dset in datasets:
    for split_type in splits:
        
        folder = "outputs/PCA/"+template.format(dset, split_type)
        for fname in os.listdir(folder):
            if not fname.startswith("test_metrics"):
                continue
            # seed = None
            # drug_name = None
            with open(os.path.join(folder, fname), "r") as f:
                m = json.load(f)
                
            m["dataset"] = dset
            m["split_type"] = split_type
            joint_pca_results.append(m)
            

    
joint_pca_results = pd.DataFrame(joint_pca_results)
joint_pca_results
```

```python
joint_pca_results = joint_pca_results.drop("n_pca_components", axis=1)
joint_pca_results = joint_pca_results[["dataset", "split_type", "AUPRC", "balanced_accuracy", "mcc"]]
joint_pca_results
```

```python
joint_pca_results.groupby(["dataset", "split_type"]).mean()
```

# Split PCA

```python
template = "SplitPCA_DRIAMS-{}_{}_metrics"


split_pca_results = []
for dset in datasets:
    for split_type in splits:
        
        folder = "outputs/PCA/"+template.format(dset, split_type)
        for fname in os.listdir(folder):
            if not fname.startswith("test_metrics"):
                continue
            # seed = None
            # drug_name = None
            with open(os.path.join(folder, fname), "r") as f:
                m = json.load(f)
                
            m["dataset"] = dset
            m["split_type"] = split_type
            split_pca_results.append(m)
            

    
split_pca_results = pd.DataFrame(split_pca_results)
split_pca_results
```

```python
# split_pca_results = split_pca_results.drop("n_pca_components", axis=1)
split_pca_results = split_pca_results[["dataset", "split_type", "AUPRC", "balanced_accuracy", "mcc"]]
split_pca_results
```

```python
df1 = split_pca_results.groupby(["dataset", "split_type"]).mean()
df2 = split_pca_results.groupby(["dataset", "split_type"]).std()
```

```python
df1.columns = [c+"_average" for c in df1.columns]
df2.columns = [c+"_std" for c in df2.columns]

cols = []
for i in range(len(df1.columns)):
    cols.append(df1.columns[i])
    cols.append(df2.columns[i])
            
               
df3 = pd.merge(df1, df2, left_index=True, right_index=True)
df3 = df3[cols]
```

```python
df3.to_csv("outputs/aggregate_resuts/SplitPCA_metrics.csv")
```
